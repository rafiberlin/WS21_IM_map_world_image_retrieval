# Game Metrics Baseline Model (Sentence Bert and Captioning)

```python
import pandas as pd
from avatar_sgg.image_retrieval.evaluation import read_game_logs, merge_log_results_in_directory

def print_metrics(log_path):
    df = merge_log_results_in_directory(log_path)
    #df[df["player_room_disclosed"] == True]
    success_rate = df[df["game_won"] == True]["game_won"].count() / df["game_won"].count()
    average_candidate_similarity =  df[df["current_candidate_similarity"] > 0 ]["current_candidate_similarity"].mean()
    print("Success Rate:", success_rate)
    print("Average Number of Interaction:", df["number_of_interaction"].mean())
    print("Average Candidate Similarity:", average_candidate_similarity)
    return df

log_path = "/home/rafi/PycharmProjects/clp-sose21-pm-vision/results/slurk_logs/baseline/"
df_baseline = print_metrics(log_path)
```
